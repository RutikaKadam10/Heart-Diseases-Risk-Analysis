```python
#Drawing correlation matrix
plt.figure(figsize=(15,6))
sns.heatmap(hd_data.corr(),annot=True)
```

###### Useful insights from above Correlations matrix -
Strong positive correlations -
1.cp ie. Chest pain type and target - shows 0.43 significant positive correlation, suggests Chest pain type, cp is imporatant predictor for the target. Means more severe chest pain results in more chances for having heart diseases.
2.thalach and target - shows 0.42 significant positive correlation, suggests heart rate, is imporatant predictor for the target.
More the heart rate during stress test, more chances of having heart diseases.thalach ie. maximum heart rate achieved by a patient during a stress test or exercise test.
3.Slope and target - shows 0.34, moderate positive correlation with the target. This feature could be relevant in prediction.
slope - Slope of the Peak Exercise ST Segment.

Strong negative correlations -
1.exang and target - shows - -0.44 negatively correlated with the target. Patients who experience angina during exercise might have a lower probability of heart disease based on the dataset.
2.ca ie. Number of Major Vessels Colored by Fluoroscopy and target - shows -0.41, negatively correlated with the target. The number of major vessels is negatively correlated with the target. Fewer major vessels could indicate a higher likelihood of heart disease.
3.Oldpeak ie. ST Depression Induced by Exercise and target - shows -0.43. Higher ST depression is negatively correlated with the target. It suggests that this feature might be important in identifying patients without heart disease.

Other insights - 
1.thalach and cp: 0.29, maximum heart rate and chest pain type are moderately positively correlated, suggesting some shared relationship.
2.age and trestbps (Resting Blood Pressure): 0.28, age and resting blood pressure have a weak positive correlation, which is expected as blood pressure tends to increase with age.
3.age and thalach: -0.4, age is negatively correlated with maximum heart rate, which aligns with the physiological trend that maximum heart rate decreases with age.

```python
#checking number of people having heart diseases and number having no diseases
#hd_data['target'].value_counts();
print('Number of people having Heart diseases: ',hd_data['target'].value_counts().get(1))
print('Number of people not having Heart diseases: ',hd_data['target'].value_counts().get(0))
```

```python
#visualizing above
plt.figure(figsize=(5, 3))
#sns.countplot(hd_data['target'])
sns.countplot(data=hd_data, x='target', palette='viridis')
plt.title('Count of Individuals by Heart Disease Prevalence')
#plt.xlabel("Heart Disease Status")
#plt.ylabel("Count")
plt.xticks(ticks=[0, 1], labels=['No Heart Disease', 'Heart Disease'])
plt.show()
```

```python
#Counting number of male and females in this dataset
#hd_data['sex'].value_counts()
print('Number of males: ',hd_data['sex'].value_counts().get(1))
print('Number of females: ',hd_data['sex'].value_counts().get(0))
```

```python
#visualizing it
plt.figure(figsize=(5,3))
sns.countplot(data=hd_data, x='sex', palette='viridis')
plt.title('Gender Distribution in Dataset')
plt.xticks([0,1],['Females','Males'])
plt.show()
```

```python
#checking gender distribution according to the target value
plt.figure(figsize=(5,3))
sns.countplot(data=hd_data, x='sex',hue='target',palette='viridis')
plt.xticks([0,1],['Females','Males'])
plt.legend(labels=['No Heart Diseases','Heart Diseases'])
plt.show()
```

```python
#checking Chest Pain Type - which pain type is more common between people
plt.figure(figsize=(7,4))
sns.countplot(data=hd_data,x='cp',palette='viridis')
plt.title('Chest Pain type Distribution')
plt.xticks([0,1,2,3],['Typical Angina','Atypical Angina','Non-Anginal Pain','Asymptomatic'],rotation=0)
plt.show()
```

```python
#Chest Pain Distribution As Per Target Variable
plt.figure(figsize=(7,4))
sns.countplot(data=hd_data,x='cp',hue='target',palette='viridis')
plt.title('Chest pain distribution as per Target')
plt.xticks([0,1,2,3],['Typical Angina','Atypical Angina','Non-Anginal Pain','Asymptomatic'])
plt.legend(labels=['No Heart Diseases','Heart Diseases'])
plt.show()
```

```python
#Fasting Blood Sugar Distribution According To Target Variable
plt.figure(figsize=(7,4))
sns.countplot(data=hd_data,x='fbs',hue='target',palette='viridis')
plt.title('Fasting Blood Sugar distribution as per Target')
plt.xticks([0,1],['fbs 120mg/dL or less','fbs greater 120mg/dL'])
plt.legend(labels=['No Heart Diseases','Heart Diseases'])
plt.show()
```

```python
#Checking Resting Blood Pressure Distribution
plt.figure(figsize=(6,3))
sns.histplot(data=hd_data,x='trestbps',color='#09649d',bins=15)
plt.title('Resting Blood Pressure Distribution')
plt.show()
```

```python
#Comparing Resting Blood Pressure As Per Sex Column
sns.FacetGrid(hd_data, hue='sex',aspect=2,palette='viridis').map(sns.kdeplot, 'trestbps',shade=True)
plt.title('Resting Blood Pressure As Per Sex Column')
plt.legend(labels=['Male','Female'])
plt.grid()
plt.show()
```
